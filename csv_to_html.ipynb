# Exporting CSV data to HTML format

```python
import pandas as pd
```

```python
# read in csv
df = pd.read_csv("Resources/weather_by_city.csv")

df.head()
```

```python
# drop extra index row
df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python
# perform export to html file
df.to_html("Resources/city_data_html.html")
```
